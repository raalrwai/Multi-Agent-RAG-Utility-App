```python
import pandas as pd
from dotenv import load_dotenv
from openai import OpenAI
from datasets import Dataset
```

```python
load_dotenv()
```

```python
import os
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
from ragas.metrics import (
    answer_relevancy,
    faithfulness,
    context_recall,
    context_precision
)
```

```python
d=os.path.join(os.getcwd(), 'tests/billing_agent/ragas_ready.csv')
df=pd.read_csv(d, encoding='ISO-8859-1')
#pd.read_csv(r"C:\Users\Premnand Dareddy\Downloads\ragas_ready.csv", encoding="ISO-8859-1")
```

```python
df.head()
```

```python
import ast
df['contexts'] = df['contexts'].apply(ast.literal_eval)
```

```python
df1 = Dataset.from_pandas(df)
```

```python
from ragas import evaluate
result = evaluate(
    df1,
    metrics=[
        context_precision,
        faithfulness,
        answer_relevancy,
        context_recall
    ],
)
result
```
